## Visualization

This notebook contains the code for visualizing the data after formulating questions and hypotheses, and doing some exploratory data analysis on the dataframe.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import pandas_profiling
import urllib.request
import datetime
import plotly.express as px
import plotly.graph_objects as go
import re
```

```python
df = pd.read_csv('clean_columns_v7.csv')
```

```python
shark_bar_01 = df.groupby(["Hemisphere", "Sex"])["Sex"].count().unstack().plot(kind="bar")
```

```python
sns.countplot(x='Activity', data=df);
```

```python
sns.countplot(x='Activity', data=df);
```

```python
sns.boxplot(x='Year', data=df);
```

```python
sns.countplot(x='Year', data=df);
```

```python
sns.violinplot(x='Sex', y='Age', data=df);
```

```python
sns.histplot(y=df['Age'], x=df['Year'], bins=5)
```

```python
activity_counts = df['Activity'].value_counts()
sns.countplot(y='Activity', data=df, order=activity_counts.index)
```

```python
hemisphere_counts = df['Hemisphere'].value_counts().sort_index()
sns.barplot(x=hemisphere_counts.index, y=hemisphere_counts.values)
```

```python
fig = px.scatter(df, x=df['Year'], y=df['Age'], animation_frame=df['Hemisphere'],
           color=df['Country'], hover_name=df['Activity'],
           log_x=True, size_max=15, range_x=[1920,2020], range_y=[1,75])
#
fig["layout"].pop("updatemenus") # optional, drop animation buttons
fig.show()
```

```python
fig = px.scatter(df, x=df['Year'], y=df['Age'], color=df['Fatal(y/n)'])
fig.show()
```

```python
shark_bar_01 = df.groupby(["Hemisphere", "Sex"])["Sex"].count().unstack().plot(kind="bar")
```

```python
top_10_countries = list(dict(df['Country'].value_counts().head(10)).keys())
df = df[df['Country'].isin(top_10_countries)]


fig = px.scatter(df, x=df['Year'], y=df['Age'], color=df['Country'], hover_name=df['Activity'],
           log_x=True, size_max=15, range_x=[1920,2020], range_y=[1,75])

fig["layout"].pop("updatemenus") # optional, drop animation buttons
fig.show()
```
